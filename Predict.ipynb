```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the data set
data = pd.read_csv("/content/dataset_clean (3).csv",encoding="utf-8")
#create a dataframe
df1=pd.DataFrame(data)
df2=df1.append(df1,ignore_index=True)
df=df2.append(df2,ignore_index=True)
print(df)
```

```python
# define the depenedent and independent variable
X = df['Symptoms']
y = df['Disease']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
x = count_vect.fit_transform(X) #Sparse matrix
print('\nTotal features extracted using CountVectorizer:',x)
print('\nFeatures for first 5 training instances are listed below')
df=pd.DataFrame(x.toarray(),columns=count_vect.get_feature_names_out())
print(df[0:5])
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sd = StandardScaler(with_mean=False)
X_train= sd.fit(x).transform(x.astype(float))
print(X_train)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_train,y)
print('\nDataset is split into Training and Testing samples')
print('Total training instances :', xtrain.shape[0])
print('Total testing instances :', xtest.shape[0])

print('Total training instances :', ytrain.shape[0])
print('Total testing instances :', ytest.shape[0])
```

```python
#Naive Bayes Text Classifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
clf = MultinomialNB().fit(xtrain,ytrain)
predict_value = clf.predict(xtrain)
# print(predict_value[1],ytrain[1])
print(ytrain)
print(predict_value)

print("The accuracy is : ",accuracy_score(ytrain,predict_value)*100)
```

```python
#KNN algorithm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
KNN = KNeighborsClassifier()
KNN.fit(xtrain,ytrain)
predict_value = KNN.predict(xtrain)
print("The accuracy is : ",accuracy_score(ytrain,predict_value)*100)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier as DT
from sklearn.metrics import accuracy_score
dt = DT()
dt.fit(xtrain,ytrain)
predict_value = dt.predict(xtrain)
print("The accuracy is : ",accuracy_score(ytrain,predict_value)*100)
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier as DT
from sklearn.metrics import accuracy_score,confusion_matrix

dt = DT()
dt.fit(xtrain,ytrain)
predict_value = dt.predict(xtrain)
print("The accuracy is : ",accuracy_score(ytrain,predict_value)*100)

cm = confusion_matrix(ytrain,predict_value)
print(cm)

#plot
fig = plt.figure(figsize=(16, 16))
sns.heatmap(cm, annot=True, fmt="d", cmap='Blues', linewidths=10)
# Add labels and color bar
plt.title('Confusion Matrix Heatmap (Training Set)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.savefig('plot.png',dpi=1200)
plt.close(fig)
print(len(cm))
```

```python
#Neural Network
from sklearn.neural_network import MLPClassifier as DT
from sklearn.metrics import accuracy_score
dt = DT()
dt.fit(xtrain,ytrain)
predict_value = dt.predict(xtrain)
print("The accuracy is : ",accuracy_score(ytrain,predict_value)*100)
```

```python
#Support Vector Classifier
from sklearn.svm import SVC as DT
from sklearn.metrics import accuracy_score
dt = DT()
dt.fit(xtrain,ytrain)
predict_value = dt.predict(xtrain)
print(predict_value[2])
print("The accuracy is : ",accuracy_score(ytrain,predict_value)*100)
```
